```python
import os

# List all subdirectories using os.listdir
basepath = r"C:\Users\nawaf alharbi\Documents\HGG"
HGG=[]
for entry in os.listdir(basepath):
    if os.path.isdir(os.path.join(basepath, entry)):
        HGG.append(entry)
```

```python
T1=[]
rootdir=os.walk(r"C:\Users\nawaf alharbi\Documents\HGG")
for root,subd,fils in rootdir:
    for name in fils:
        if name.endswith('t1.nii.gz'):
            T1.append(os.path.join(root,name)) 
```
